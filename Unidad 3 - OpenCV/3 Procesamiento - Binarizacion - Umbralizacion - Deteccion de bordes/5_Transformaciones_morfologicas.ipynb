### Transformaciones morfol√≥gicas
Las transformaciones morfol√≥gicas son algunas operaciones simples basadas en la forma de la imagen, que normalmente se aplican a im√°genes binarias. 

La transformaci√≥n morfol√≥gica, como su nombre lo dice, transforma o modifica la forma de la imagen basandose en un elemento de an√°lisis o kernel. 

A continuaci√≥n se presentan las tranformaciones morfol√≥gicas m√°s comunes.

### Erosi√≥n 
La idea b√°sica de la erosi√≥n es similar al concepto de erosi√≥n del suelo, erosiona los l√≠mites del objeto de primer plano (si este es el que se tiene en blanco). 

El kernel se desliza a trav√©s de la imagen (como en la convoluci√≥n 2D). 

- Un p√≠xel en la imagen original (1 o 0) se considerar√° 1 solo si todos los p√≠xeles debajo del kernel son 1, de lo contrario se erosiona, es decir, se cambia su valor a cero. 

Entonces, lo que sucede es que todos los p√≠xeles cerca del l√≠mite se descartar√°n dependiendo del tama√±o del kernel. 

Por lo tanto, el grosor o el tama√±o del objeto en primer plano disminuye o simplemente la regi√≥n blanca disminuye en la imagen. 

Es √∫til para eliminar peque√±os ruidos blancos.

```
cv2.erode(imgB,kernel,iterations = 1)
    imgB       : Imagen de preferencia binaria.
    kernel     : Arreglo de unos con el tama√±o de la transformaci√≥n deseada.
    iterations : N√∫mero de veces que se desea repetir este proceso.
```

````python
kernel = np.ones((13,13),np.uint8)
kernel
````

````python
imaEro = cv2.erode(imgB,kernel,iterations = 1)
grafica(imgB, imaEro)
````

### Dilataci√≥n
Es justo lo contrario al proceso de la erosi√≥n. 

- Un elemento de p√≠xel es ‚Äò1‚Äô si al menos un p√≠xel debajo del n√∫cleo es ‚Äò1‚Äô. 

Por lo tanto, esto hace que aumente la regi√≥n blanca en la imagen o el tama√±o del objeto en primer plano. 

En algunos casos, la dilataci√≥n es utilizada para eliminar el ruido negro al interior de nuestro elemento de inter√©s. 

````python
imaDil = cv2.dilate(imgB,kernel,iterations = 1)
grafica(imgB, imaDil)
````

### Apertura
En algunos caso cuando se elimina el ruido blanco mediante la erosi√≥n, nuestros elementos de inter√©s pierden parte del tama√±o y por lo tanto, parte de informaci√≥n de int√©res. 

En este caso, se puede proceder con una dilataci√≥n ulterior con el fin de recuperar el tama√±o original. 

Este proceso es conocido como apertura. 

En opencv podemos implementar la apertura con los dos comandos anteriores o mediante la siguiente funci√≥n:

```
imaOp2=cv2.morphologyEx(imgB, Trans, kernel, iterations = 1)
    imgB       : Imagen de preferencia binaria.
    Transf     : Tipo de transformaci√≥n morfol√≥gica
            cv2.MORPH_OPEN
            cv2.MORPH_CLOSE
            cv2.MORPH_GRADIENT
            cv2.MORPH_TOPHAT
            cv2.MORPH_BLACKHAT
    kernel     : Arreglo de unos con el tama√±o de la transformaci√≥n deseada.
    iterations : N√∫mero de veces que se desea repetir este proceso.
```

````python
imaEro = cv2.erode(imgB,kernel,iterations = 1)
imaOp1 = cv2.dilate(imaEro,kernel,iterations = 1)
imaOp2=cv2.morphologyEx(imgB, cv2.MORPH_OPEN, kernel, iterations = 1)
grafica(imgB, imaOp1)
grafica(imgB, imaOp2)
````

### Cierre
El cierre es al rev√©s de Apertura, Dilataci√≥n seguida de Erosi√≥n. 

Es √∫til para cerrar peque√±os orificios dentro de los objetos en primer plano o peque√±os puntos negros en el objeto.

````python
imaEro = cv2.dilate(imgB,kernel,iterations = 1)
imaCl1 = cv2.erode(imaEro,kernel,iterations = 1)
imaCl2=cv2.morphologyEx(imgB, cv2.MORPH_CLOSE, kernel, iterations = 1)

grafica(imgB, imaCl1)
grafica(imgB, imaCl2)
````

### Gradiente
Es la diferencia entre dilataci√≥n y erosi√≥n de una imagen.


El resultado se ver√° como el contorno del objeto.

````python
imaT=cv2.morphologyEx(imgB, cv2.MORPH_GRADIENT, kernel, iterations = 1)
grafica(imgB, imaT)
````

### Sombrero de copa
Es la diferencia entre la imagen de entrada y la apertura de la imagen.

````python
imaT=cv2.morphologyEx(imgB, cv2.MORPH_TOPHAT, kernel, iterations = 1)
grafica(imgB, imaT)
````

### Sombrero negro
Es la diferencia entre el cierre de la imagen de entrada y la imagen de entrada.

````python
imaT=cv2.morphologyEx(imgB, cv2.MORPH_BLACKHAT, kernel, iterations = 1)
grafica(imgB, imaT)
````

### Esqueletizaci√≥n
La esqueletizaci√≥n reduce los objetos binarios a representaciones de 1 p√≠xel de ancho. 

Esto puede ser √∫til para la extracci√≥n de caracter√≠sticas y/o representar la topolog√≠a de un objeto.

La esqueletizaci√≥n funciona haciendo pases sucesivos de la imagen. En cada pasada, los p√≠xeles del borde se identifican y eliminan con la condici√≥n de que no rompan la conectividad del objeto correspondiente.

````python
skel = np.zeros(imgB.shape,np.uint8)
````

````python
element
````

````python
imgS=imgB.copy()
while(True):
    eroded = cv2.erode(imgS,element)
    temp = cv2.dilate(eroded,element)
    temp = cv2.subtract(imgS,temp)
    skel = cv2.bitwise_or(skel,temp)
    imgS = eroded.copy()

    zerosN = cv2.countNonZero(imgS)
    if zerosN==0:
        break
````

````python
grafica(imgB, skel)
````
